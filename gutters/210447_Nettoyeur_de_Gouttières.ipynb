```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
def gutter(number):
    # Load the image
    image_path = f"gutters{number}.JPG"
    image = cv2.imread(image_path)

    # Display the original image
    cv2_imshow(image)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Get the dimensions of the image
    height, width = gray.shape[:2]
    print(f"{width} x {height}")

    # Apply different operations based on the input number
    if number == 1:
        # Dilate the grayscale image
        dilated_img = cv2.dilate(gray, np.ones((5, 5)))

        # Apply Gaussian blur
        blurred_img = cv2.GaussianBlur(dilated_img, (7, 7), 128)

        # Compute the final image
        final = 255 - cv2.absdiff(blurred_img, gray)

        # Modify the first 100 pixels in each row
        for i in range(height):
            for j in range(100):
                final[i, j] = (final[i, j] / 255) ** (1.4) * 255

    elif number == 2:
        # Dilate the grayscale image
        dilated_img = cv2.dilate(gray, np.ones((19, 19)))

        # Apply median blur
        blurred_img = cv2.medianBlur(dilated_img, 23)

        # Compute the final image
        final = 255 - cv2.absdiff(blurred_img, gray)

    elif number == 3:
        # Apply adaptive thresholding
        final = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 2)


    # Display the final image
    cv2_imshow(final)
```

```python
gutter(1)
gutter(2)
gutter(3)
```
