```python
import cv2 as cv
import numpy as np
```

```python
def image_enhancer(number):

    image_path = f'cctv{number}.JPG'
    image = cv.imread(image_path)

    gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY) #conversion to grayscale

    equalized = cv.equalizeHist(gray) #histogram equalization to enhance the image

    clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8)) # Increase contrast using CLAHE
    enhanced = clahe.apply(equalized)

    enhanced_path = f'enhanced-cctv{number}.jpg'
    cv.imwrite(enhanced_path, enhanced)

    print(f'Enhanced image saved as {enhanced_path}')
```

```python
number = int(input('Enter cctv number (1, 2, 3, or 4): '))
image_enhancer(number)
```
